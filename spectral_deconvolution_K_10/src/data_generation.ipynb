```python
import numpy as np
import matplotlib.pyplot as plt
a_star = np.array([0.587,1.522,1.183,0.864,0.574,0.388,0.53,0.3,0.6,0.7])
mu_star = np.array([0.75,0.84,0.95,1.71,1.82,1.95,1.12,1.30,1.45,2.1])
b_star = np.array([500.689,1046.837,764.469,550.689,562.837,634.469,600.689,664.469,700.689,746.837])
data_num = 301
x_data = np.linspace(0,3,data_num)
def g(x_i,a,mu,b):
    ans = 0
    for i in range(len(a)):
        ans += a[i]*np.exp(-b[i]*(x_i-mu[i])**2/2)
    return ans
y_data = g(x_data,a_star,mu_star,b_star) + np.random.randn(data_num)*0.1
plt.plot(x_data,g(x_data,a_star,mu_star,b_star))
plt.scatter(x_data,y_data,s = 1)
f = open("../input/data_0.txt","w")
for i in range(data_num):
    f.write(str(y_data[i]) + "\n")
f.close()
```
