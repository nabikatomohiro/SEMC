```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special
from scipy.stats import wasserstein_distance
```

```python
data_num = 30000
ratio = 1.001
free_energy_baseline = -np.log(np.pi/data_num*special.erf(np.sqrt(data_num)/4)*(special.erf(np.sqrt(data_num*ratio)/4)/np.sqrt(ratio) + special.erf(np.sqrt(data_num)/4)*np.exp(-data_num*(ratio-1)/16)))
```

```python
folder_name_base = "../multimodal_sampling/output/EMC/"
sample_array = [300, 1000, 3000, 10000, 30000, 100000, 300000, 1000000]
L_array = [10,30,100,300]
change = 4
iter = 10
free_energy_array_EMC_multimodal = np.zeros((len(L_array),change,iter))
Time_array_EMC_multimodal = np.zeros((len(L_array),change,iter))
for i in range(len(L_array)):
    L = L_array[i]
    for j in range(change):
        folder_name = folder_name_base + "burn_in_" + str(sample_array[3-i+j]) + "_L_" + str(L_array[i]) + "/"
        for k in range(iter):
            f = open(folder_name + "trial_" + str(k) + "/free_energy.txt", "r")
            l = f.readlines()
            free_energy_array_EMC_multimodal[i,j,k] = float(l[0]) - free_energy_baseline
            f = open(folder_name + "trial_" + str(k) + "/calculation_time.txt", "r")
            l = f.readlines()
            Time_array_EMC_multimodal[i,j,k] = float(l[0])
```

```python
fig, axs = plt.subplots(4, 4, dpi=300, figsize=(20,12))
spines = 1.5
plt.rcParams['figure.subplot.bottom'] = 0.1
plt.rcParams['figure.subplot.top'] = 0.9
plt.rcParams['figure.subplot.right'] = 0.9
plt.rcParams['figure.subplot.left'] = 0.125
plt.rcParams['font.family'] = 'Times New Roman' # font family
plt.rcParams['mathtext.fontset'] = 'stix' # math font
plt.rcParams["font.size"] = 25
Title_name = ["(I) multimodal sampling", "(II - a) $K = 3$","(II - b) $K$ = 4","(III) Exhaustive Search"]
Folder_name = ["Simple","Deconvolution","Reference"]
xmin_array = [0.1,1,1.5,1]
xmax_array = [2e1, 2e2, 3e2,2e2]
ymin_array = [-0.1,-0.2,-0.2,-1]
ymax_array = [0.1,0.2,0.2,1]
# free_energy_array_EMC_array = [free_energy_array_EMC_multimodal, free_energy_array_EMC_deconvolution, free_energy_array_EMC_reference]
free_energy_array_SEMC_array = [free_energy_array_SEMC_multimodal, free_energy_array_SEMC_deconvolution, free_energy_array_SEMC_deconvolution_K_4, free_energy_array_SEMC_exhaustive]
# Time_array_EMC_array = [Time_array_EMC_multimodal, Time_array_EMC_deconvolution, Time_array_EMC_reference]
Time_array_SEMC_array = [Time_array_SEMC_multimodal, Time_array_SEMC_deconvolution, Time_array_SEMC_deconvolution_K_4, Time_array_SEMC_exhaustive]
text = ["(1) SEMC J = 0.1","(2) SEMC  J = 0.3","(3) SEMC  J = 0.7","(4) SEMC  J = 0.9"]
for iter in range(len(axs)):
    xmin = xmin_array[iter]
    xmax = xmax_array[iter]
    ymin = ymin_array[iter]
    ymax = ymax_array[iter]
    # free_energy_array_EMC = free_energy_array_EMC_array[iter]
    free_energy_array_SEMC = free_energy_array_SEMC_array[iter]
    # Time_array_EMC = Time_array_EMC_array[iter]
    Time_array_SEMC = Time_array_SEMC_array[iter]
    for i, ax in enumerate(axs[iter]):
        #0
        ax.axhline(0, color='black', linewidth=1, linestyle='--')
        ax.spines["top"].set_linewidth(spines)
        ax.spines["left"].set_linewidth(spines)
        ax.spines["bottom"].set_linewidth(spines)
        ax.spines["right"].set_linewidth(spines)
        ax.yaxis.set_ticks_position('both')
        ax.xaxis.set_ticks_position('both')
        ax.tick_params(direction = "in", length = 7)
        ax.set_xscale("log")
        ax.set_ylim([ymin,ymax])
        ax.set_xlim([xmin,xmax])
        if i < 2:
            x = Time_array_SEMC[i]
            y = free_energy_array_SEMC[i]
        else:
            x = Time_array_SEMC[i+1]
            y = free_energy_array_SEMC[i+1]
        ax.scatter(x,y,s=5,color = "blue",alpha = 0.5,label = "SEMC with Exhange Rate J")
        ax.errorbar(np.mean(x,axis = 1),np.mean(y,axis = 1),yerr = np.std(y,axis = 1),color = "blue",linewidth = 2, label = "Mean with Exhange Rate J", capsize=5)
        x = Time_array_SEMC[2]
        y = free_energy_array_SEMC[2]
        ax.scatter(x,y,s=5,color = "red",alpha = 0.5,label = "SEMC with Exhange Rate 0.5")
        ax.errorbar(np.mean(x,axis = 1),np.mean(y,axis = 1),yerr = np.std(y,axis = 1),color = "red",linewidth = 2, label = "Mean with Exhange Rate 0.5", capsize=5)
        ax.tick_params(pad = 10)
        ax.text(np.exp((np.log(xmin) + np.log(xmax))/2), ymax*1.15,text[i], ha='center')
        if iter == 0 and i == 3:
            #
            ax.legend(loc='lower right', bbox_to_anchor=(1.0, 1.3), fontsize=25,ncol = 2)
    axs[iter][0].text(xmin*0.1, ymax*1.8, Title_name[iter], ha='left', va='center')
fig.supylabel("Error",x = 0.07, y = 0.63)
fig.supxlabel("Calculation Time(s)",x = 0.5, y = 0.02)
plt.subplots_adjust(wspace=0.3, hspace=1.0)
plt.savefig("../../../MonteCarloPaper//figure/SEMC_freeenergy", bbox_inches="tight")
```
